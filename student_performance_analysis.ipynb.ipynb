# Student Performance Analysis using Python & Pandas

**Dataset:** Students Performance in Exams  
**Tools:** Python, Pandas, Matplotlib, Seaborn  

## Objective
The goal of this project is to analyze student performance data and identify patterns related to gender, test preparation, lunch type, and score distribution using basic data analytics techniques.


```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

Insights from Student Performance Analysis

Insight 1: Gender and Math Performance
The average math score differs slightly between genders. Female students have a marginally higher average math score compared to male students, indicating a small gender-based performance difference in mathematics.

Insight 2: Effect of Test Preparation
Students who completed the test preparation course performed better on average in all three subjectsmath, reading, and writingcompared to students who did not complete the course. This shows that structured preparation positively impacts academic performance.

Insight 3: Distribution of Math Scores
The histogram of math scores shows that most students fall within a middle score range, with fewer students achieving very low or very high scores. This indicates a fairly balanced distribution of math performance.
